```python

# coding: utf-8

# Name - Prajakta Pitale DrexelID - 14193330

# In[6]:


import numpy as np
import pandas as pd
class logisticRegression():
    
    def __init__(self):
        self.coef = 0
        
    def sigmoid(self, array):
        return 1 / (1 + np.exp(-array)).round(3)
        
    def fit(self, x, y, alpha, threshold):
        if x.shape[0] != y.shape[0]:
            raise ValueError("Size doesn't match!")
        else:
            pass
        
        if x.isnull().values.any() == True:
            raise ValueError("Input Data has missing values")
        else:
            pass

        if y.isnull().values.any() == True:
            raise ValueError("Target has missing values")
        else:
            pass
        
        for i in x.columns:
            if x[i].dtype == 'object':
                raise ValueError("Data has categorical variables!")
            else:
                pass
        
        if y.dtype == 'object':
                raise ValueError("Target is categorical!")
        else:
            pass
        
        x = x.as_matrix(columns=None)
        y = y.as_matrix(columns=None)
        y = np.array(y).reshape(6366L,1L)
        bias = np.ones((x.shape[0],1))
        x = np.concatenate((bias, x), axis = 1)
        coef = np.random.uniform(low=-2, high=2, size=(x.shape[1],1))
        cost = []
        N = 1
        while(N < 10000):
            coef_times_inputs = np.dot(x,coef)
            yhat = self.sigmoid(coef_times_inputs)
            error = y - yhat
            a = np.sum((-y*np.log(yhat+0.000001)) - ((1-y)*np.log(1-yhat+0.00001)))/ x.shape[0]
            cost = np.append(cost,a)
            b = abs(cost[-5:].mean() - cost[-1])
            if b < threshold and N>5:
                N = N + 1
            else:
                pass
            gradient = np.dot(x.T,error)/x.shape[0]
            coef = coef - (alpha*gradient)
            alpha = alpha * 0.95
            N = N + 1
        self.coef = coef
    
    
    def predict_prob(self, x):
        x = x.as_matrix(columns=None)
        bias = np.ones((x.shape[0],1))
        x = np.concatenate((bias, x), axis = 1)
        coef_times_inputs = np.dot(x,self.coef)
        yhat = (1/(1+np.exp(-coef_times_inputs))).round(3)      
        return yhat
    
    
    def predict_class(self, x):
        yhat = self.predict_prob(x) 
        yhat[yhat<0.5] = 0
        yhat[yhat>=0.5] = 1
        return yhat
    
    def get_accuracy(self, x, y):
        y = y.as_matrix(columns=None)
        y = np.array(y).reshape(6366L,1L)
        yhat = self.predict_class(x)
        accuracy = float(np.sum(y==yhat)/float(x.shape[0])) * 100
        print("The accuracy is:")
        return accuracy
```
